```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def init_browser():
    executable_path = {'executable_path': 'chromedriver.exe'}
    return Browser('chrome', **executable_path, headless=False)
```

```python
browser = init_browser()

#list url to visit
season = '2018-19'
season_type = 'Regular%20Season'
date_to = '02%2F18%2F2019'

nba_url = 'https://stats.nba.com/teams/boxscores/?Season={season}&SeasonType={season_type}&DateTo={date_to}'.format(season=season, season_type=season_type, date_to=date_to)

#visit url
browser.visit(nba_url)
print(nba_url)
```

```python
full = browser.find_by_tag('nba-stat-table')
full_t = full.find_by_tag('select')[0]
full_table = full_t.find_by_tag('option')[0]
full_table.click()
```

```python
nba_html = browser.html 
soup = BeautifulSoup(nba_html, 'html.parser')
```

```python
table = soup.find_all('nba-stat-table')[0]
```

```python
nba_stats = pd.read_html(str(table), header=0)[0]
nba_stats
```

```python
nba_stats.to_csv('Raw_Compiled_Stats_1819.csv')
```

```python
nba_stats.columns
```

```python
nba_stats.rename(columns = {list(nba_stats)[2]: 'Game Date'}, inplace = True)
```

```python
nba_stats['Game Date'] = pd.to_datetime(nba_stats['Game Date'])
nba_stats.head()
```

```python
nba = nba_stats[['Team', 'Game Date', 'PTS', 'FG%', '3P%', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'BLK', 'TOV', 'PF']]
nba.head()
```

```python
avg_1819 = nba.groupby(['Team']).mean()
avg_1819.head()
```

```python
avg_1819.reset_index(inplace=True)
avg_1819.head()
```

```python
avg_1819.columns
```

```python
avg_1819.rename(columns = {list(avg_1819)[0]: 'TEAM'}, inplace = True)
avg_1819.head()
```

```python
stats_1819 = avg_1819.drop(labels='TEAM', axis=1)
stats_1819.head()
```

```python
diff = []

for x in range(1, 30):
    diff_1819 = stats_1819 - stats_1819.shift(periods =- x)
    diff_1819.drop(diff_1819.tail(x).index, inplace=True)
    home_team = avg_1819.iloc[:,0]
    home_team.drop(home_team.tail(x).index, inplace=True)
    away_team = avg_1819.iloc[:,0].shift(periods =- x)
    away_team.drop(away_team.tail(x).index, inplace=True)
    diff_1819['AWAY'] = away_team
    diff_1819['HOME'] = home_team
    diff.append(diff_1819)

diff_reg_1819 = pd.concat(diff, axis=0, sort=True)
diff_reg_1819.head()
```

```python
matchup_1819 = diff_reg_1819[['HOME', 'AWAY', 'PTS', 'FG%', '3P%', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'BLK', 'TOV', 'PF']]
matchup_1819.head()
```

```python
neg_matchup_1819 = matchup_1819.iloc[:,2:13].multiply(-1)
neg_matchup_1819.head()
```

```python
neg_matchup_1819['HOME'] = matchup_1819['AWAY']
neg_matchup_1819['AWAY'] = matchup_1819['HOME']
neg_matchup_1819.head()
```

```python
neg_matchup_1819 = neg_matchup_1819[['HOME', 'AWAY', 'PTS', 'FG%', '3P%', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'BLK', 'TOV', 'PF']]
neg_matchup_1819.head()
```

```python
full_1819 = pd.concat([matchup_1819, neg_matchup_1819], axis=0, ignore_index=True, sort=True)
full_1819.head()
```

```python
full_1819 = full_1819[['HOME', 'AWAY', 'PTS', 'FG%', '3P%', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'BLK', 'TOV', 'PF']]
full_1819.head()
```

```python
full_1819.to_csv('Plus_Minus_1819.csv')
```

```python
cha_hou = full_1819.loc[(full_1819['HOME'] == 'CHA') & (full_1819['AWAY'] == 'HOU')]
atl_min = full_1819.loc[(full_1819['HOME'] == 'ATL') & (full_1819['AWAY'] == 'MIN')]
bkn_was = full_1819.loc[(full_1819['HOME'] == 'BKN') & (full_1819['AWAY'] == 'WAS')]
mia_gsw = full_1819.loc[(full_1819['HOME'] == 'MIA') & (full_1819['AWAY'] == 'GSW')]
bos_por = full_1819.loc[(full_1819['HOME'] == 'BOS') & (full_1819['AWAY'] == 'POR')]
mem_chi = full_1819.loc[(full_1819['HOME'] == 'MEM') & (full_1819['AWAY'] == 'CHI')]
sas_det = full_1819.loc[(full_1819['HOME'] == 'SAS') & (full_1819['AWAY'] == 'DET')]
dal_ind = full_1819.loc[(full_1819['HOME'] == 'DAL') & (full_1819['AWAY'] == 'IND')]
uta_lac = full_1819.loc[(full_1819['HOME'] == 'UTA') & (full_1819['AWAY'] == 'LAC')]
sac_mil = full_1819.loc[(full_1819['HOME'] == 'SAC') & (full_1819['AWAY'] == 'MIL')]
lal_nop = full_1819.loc[(full_1819['HOME'] == 'LAL') & (full_1819['AWAY'] == 'NOP')]

predict_1819 = pd.concat([cha_hou, atl_min, bkn_was, mia_gsw, bos_por,
                          mem_chi, sas_det, dal_ind, uta_lac,
                          sac_mil, lal_nop], axis=0, sort=True)

predict_1819 = predict_1819[['HOME', 'AWAY', 'PTS', 'FG%', '3P%', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'BLK', 'TOV', 'PF']]
predict_1819.reset_index(drop=True, inplace=True)
predict_1819
```

```python
X_1819 = predict_1819.drop(['HOME', 'AWAY', 'PTS'], axis=1)
```

```python
import pickle
filename = 'NBA_Prediction.sav'

loaded_model = pickle.load(open(filename, 'rb'))
```

```python
predictions_1819 = loaded_model.predict(X_1819)
predictions_1819
```

```python
score_prediction_1819 = pd.DataFrame(data=predictions_1819)
```

```python
game_predict_1819 = pd.concat([predict_1819, score_prediction_1819], axis=1, sort=True)
game_predict_1819
```

```python
game_predict_1819.rename(columns = {list(game_predict_1819)[13]: 'Score Prediction'}, inplace = True)

game_predict_1819
```

```python
presentation_prediction = game_predict_1819[['HOME', 'AWAY', 'Score Prediction']]
presentation_prediction
```
